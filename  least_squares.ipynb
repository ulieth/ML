```python
X = data_series.index.to_numpy()
Y = data_series['y'].to_numpy()
X
```

```python
plt.figure(figsize=(20,8))
plt.plot(Y)
plt.xlabel('Time step')
plt.ylabel('Y')
plt.show()
```

```python
N = data_series.shape[0]
N
```

```python
a = (N * (X*Y).sum() - X.sum()*Y.sum())/(N*(X*X).sum()-X.sum()**2) # => or (N * np.dot(X,Y) - X.sum()*Y.sum())/(N*np.dot(X,X)-X.sum()**2)
b = (Y.sum()-a*X.sum())/N
print(a,b)
```

```python
x = np.arange(1,101)
data_series['F_MMC'] = a*x+b
data_series
```

```python
plt.figure(figsize = (20,8))
plt.plot('y', data=data_series)
plt.plot('F_MMC',data=data_series)
plt.xlabel('Time step')
plt.ylabel('Y')
plt.show()
```
